```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Load the .npy file
#file_path = np.load('/content/Converted_data/testing/heartwood/cuboid_11284.npy')
data = np.load('/content/Converted_data/testing/heartwood/cuboid_11284.npy')


# Check the shape of the data
print("Original shape:", data.shape)

# Reshape the data to (number_of_images, image_size)
# In this case: (320, 32*32)
reshaped_data = data.reshape(-1, 32*32)
print("Reshaped data shape:", reshaped_data.shape)

# Apply PCA
pca = PCA(n_components=10)  # Choose the number of components
pca_result = pca.fit_transform(reshaped_data)
print("PCA result shape:", pca_result.shape)

# Inverse transform to reconstruct images from PCA components
reconstructed_data = pca.inverse_transform(pca_result)
print("Reconstructed data shape:", reconstructed_data.shape)

# Reshape the first reconstructed image back to the original shape
first_reconstructed_image = reconstructed_data[0].reshape(32, 32)
plt.imshow(first_reconstructed_image, cmap='gray')
plt.title('First Reconstructed Image from PCA')
plt.show()

# Optional: Visualize the first principal component's influence
principal_component_image = pca.components_[0].reshape(32, 32)
plt.imshow(principal_component_image, cmap='gray')
plt.title('First Principal Component')
plt.show()
```

```python
import numpy as np
from pca import pca
import matplotlib.pyplot as plt

# Load your data
data = np.load('/content/Converted_data/testing/heartwood/cuboid_11284.npy')

# Check the shape of the data
print("Original shape:", data.shape)

# Reshape the data to (number_of_images, image_size)
reshaped_data = data.reshape(-1, 32*32)
print("Reshaped data shape:", reshaped_data.shape)

# Initialize PCA
model = pca(n_components=10)  # Specify the number of principal components
results = model.fit_transform(reshaped_data)

# The results object contains the transformed data and various plots
print("PCA results shape:", results['PC'].shape)

# Plot explained variance
model.plot()
plt.show()

# Create a biplot with specified jitter argument
model.biplot(jitter=0.1)
plt.show()

# Reconstruct the first image using the PCA components
principal_components = results['PC']
components = results['loadings']

# Transform back to the original space
reconstructed_data = np.dot(principal_components, components.T) + model.m[0]

# Reshape the first reconstructed image back to the original shape
first_reconstructed_image = reconstructed_data[0].reshape(32, 32)
plt.imshow(first_reconstructed_image, cmap='gray')
plt.title('First Reconstructed Image from PCA')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy file
ntt = np.load('/content/Converted_data/testing/heartwood/cuboid_11284.npy')

# Check the shape to understand the dimensions
print(ntt.shape)

# Function to display the image and capture click events
def onclick(event):
    # Get the x and y coordinates of the click
    x = int(event.xdata)
    y = int(event.ydata)

    # Get the pixel value at the clicked coordinates
    pixel_value = ntt[y, x, 14]

    # Print the coordinates and pixel value
    print(f'Coordinates: (x={x}, y={y})')
    print(f'Pixel value: {pixel_value}')

# Plot a single image slice (e.g., the 15th slice)
fig, ax = plt.subplots()
ax.imshow(ntt[:, :, 14], cmap='gray')
plt.title('15th Image Slice')

# Connect the click event to the handler
cid = fig.canvas.mpl_connect('button_press_event', onclick)

plt.show()
```
